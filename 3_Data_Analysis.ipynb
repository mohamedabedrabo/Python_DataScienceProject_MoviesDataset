## Discover the best and the worst movies 

```python
# this library will help us present the data in a modern way:
from IPython.display import HTML
```

```python
# now we create a subset of df , most important columns :
df_best = df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity" ]].copy()
df_best
```

```python
# we add to important features: profit = revenue-budget, ROI = revenue\budget:
df_best["profit_musd"] = df.revenue_musd.sub(df.budget_musd)
df_best["return"] = df.revenue_musd.div(df.budget_musd)
```

```python
df_best
```

```python
# we change the labels for data presentation purposes :
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes",
               "Average_Rating", "Popularity", "Profit", "ROI"]
```

```python
# now we make "Title" the main index :
df_best.set_index("Title", inplace=True)
```

```python
df_best
```

```python
# we can see that that the second column contains image (HTML code)
# so let's look at the first field :
# iloc [row,column]
df_best.iloc[0,0]
```

```python
# let's create a small df from aour original df :
subset = df_best.iloc[:5,:2]
subset
```

```python
# convert the df into HTML :
subset.to_html(escape=False)
```

```python
# like this we can present it in a nice HTML way :
HTML(subset.to_html(escape=False))
```

```python
# we want to sort the movies by their rate :
# but we can also see that the highest rated movies are movies with only 1 vote so its not really realistic.
df_best.sort_values(by = "Average_Rating", ascending=False)
```

```python
# the same problem we face with ROI, we see high ROI because the budget was 0 :
df_best.sort_values(by = "ROI", ascending = False)
```

```python
# so now we add functionality that the budget must be > 5M :
# loc is used for real values not places.
# so that result makes more sense, we can do it with 10M or like we see best:
df_best.loc[df_best.Budget >= 5].sort_values(by = "ROI", ascending = False)
```

```python
# something bery important, before we perform a filter we need to take care
# of NA (missing values) : 
df_best.Budget.fillna(0, inplace = True)
df_best.Votes.fillna(0, inplace = True)
```

```python
df_best.info()
```

```python
# now we create a function that best and worst movies:
# n is the number of rows we want to present, by is by what value we want to do the filter
# the () & () inside the loc() are our conditions for what we want to present
# ["",] is what we want to present after filtereng the df ("" wich is the first row(pictures)
# ans the "by" that we choose (like "profit" or whatever we want)):
def best_worst(n, by, ascending = False, min_bud = 0, min_votes = 0):
    
    df2 = df_best.loc[(df_best.Budget >= min_bud) & (df_best.Votes >= min_votes),
                     ["",by]].sort_values(by = by, ascending = ascending).head(n).copy()
    
    return HTML(df2.to_html(escape=False))
```

### Movies Top 5 - Highest revenue : 

```python
best_worst(5, by = "Revenue")
```

### Movies Top 5 - Highest budget : 

```python
best_worst(5, by = "Budget")
```

### Movies Top 5 - Highest profit : 

```python
best_worst(5, by = "Profit")
```

### Movies Top 5 - Lowest profit : 

```python
best_worst(5, by = "Profit", ascending = True)
```

### Movies Top 5 - Highest ROI : 

```python
# if we chose 0 as min budget we will have movies who have budget of 0
# and so there ROI would be so hight and it's not realistic so we choose 5
best_worst(5, by = "ROI" , min_bud = 5)
```

### Movies Top 5 - Lowest ROI : 

```python
#less than 50 are all 0 movies , we want to see more
# we can play with it, try 5,10,20,50 untill we see result that we want:
best_worst(5, "ROI", ascending = True, min_bud = 50)
```

### Movies Top 5 - Most votes : 

```python
best_worst(5, "Votes")
```

### Movies Top 5 - Highest Rating : 

```python
best_worst(5, "Average_Rating", min_votes = 2000)
```

### Movies Top 5 - Lowest Rating : 

```python
best_worst(5, "Average_Rating",ascending = True, min_votes = 2000, min_bud = 5)
```

```python
best_worst(5,"Popularity")
```

# Are Franchises (series of movies) more succesful ?

```python
df.head()
```

```python
# we can see here that the NaN values says that the movie doesn't belong to a franchise :
df.belongs_to_collection
```

```python
# so we create a new boolean column with 0,1 if the movie belongs to franchise :
df["Franchise"] = df.belongs_to_collection.notna()
```

```python
df.Franchise
```

```python
df.Franchise.value_counts()
```

### Franchise vs. Stand-alone Average revenue :

```python
# we use the method that looks like SQL (group by):
# df.froupby(what we want to groupy by).values we want to presetnt.functionality(mean,median,...):
# we can see that revenue of franchises are higher
df.groupby("Franchise").revenue_musd.mean()
```

### Franchise vs. Stand-alone ROI :

```python
# we must add the ROI colum to df:
df["ROI"] = df.revenue_musd.div(df.budget_musd)
```

```python
# then apply the group by method :
# in ROI it's more logical and accurate to use the median not mean :
# we can see that franchise returns 3.71 times of the budget
df.groupby("Franchise").ROI.median()
```

```python
df.head()
```

### Franchise vs. Stand-alone average budget :

```python
# we can see clearly that franchises cost more:
df.groupby("Franchise").budget_musd.mean()
```

### Franchise vs. Stand-alone average popularity :

```python
df.groupby("Franchise").popularity.mean()
```

### Franchise vs. Stand-alone average rating :

```python
# we can see that ratings are so close (the same)
df.groupby("Franchise").vote_average.mean()
```

### Franchise vs. Stand-alone SUMMARY:

```python
# we can sum everyhting in one table using agg function :
df.groupby("Franchise").agg({"budget_musd":"mean", "revenue_musd":"mean", "vote_average":"mean",
                            "popularity":"mean", "ROI":"median", "vote_count":"mean"})
```

# Most succesful Franchises : 

```python
df.belongs_to_collection
```

```python
# take a look about the values in theis column :
df.belongs_to_collection.value_counts()
```

```python
# we want to look at the values of each franchise
# when we do "budget":["sum":"mean"] then we want to see both values (sum and mean)
# so we will see 2 columns under budget for example (sum , mean):

franchises = df.groupby("belongs_to_collection").agg({"title":"count", "budget_musd":["sum","mean"],
                                                     "revenue_musd":["sum","mean"],
                                                     "vote_average":"mean", "popularity":"mean",
                                                     "ROI":"median", "vote_count":"mean"})
franchises
```

```python
# like this we we can pull the 20 largest count's in term of the title column :
# we can see that Bowery boys collection has 29 movies!! , James bond has 26.
franchises.nlargest(20, ("title", "count") )
```

```python
# The highest 20 revenue franchises by the "sum" - total budget they spent on all movies:
# we can see that harry potter collection spent 7707M$.
franchises.nlargest(20, ("revenue_musd", "sum") )
```

```python
# we can also filter by the "mean" revenue - average spending on each movie :
# Avatar is leading with 2787M$
franchises.nlargest(20, ("revenue_musd", "mean") )
```

```python
# Most expensive in terms of total budget :
# James bond leading with 1539 M$
franchises.nlargest(20, ("budget_musd", "sum"))
```

```python
# like this we can filter with vote counts that have mean >1000 :
# and then we choose the highest 20 means : 
franchises[franchises[("vote_count", "mean")] >= 1000].nlargest(20,("vote_average", "mean"))
```
