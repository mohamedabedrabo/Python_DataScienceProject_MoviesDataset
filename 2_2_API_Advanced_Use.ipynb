```python
import pandas as pd
import requests
pd.options.display.max_columns = 30
```

### "Discover" module

```python
# we can find the method on the movies API documentation:
# it's : url + /discover + /
discover_api = "https://api.themoviedb.org/3/discover/movie?"
```

```python
# and then we set our API key that we got from the movies website :
api_key = "api_key=XXXXXX"
```

```python
# we want to the movies between 01/01-29/02 of 2020
# so we use this query also from the documentation of the website :
query = "&primary_release_date.gte=2020-01-01&primary_release_date.lte=2020-02-29"
```

```python
# so we create our final url :
url = discover_api + api_key + query
#url
```

```python
# convert the data into a JSON :
data = requests.get(url).json()
```

```python
data
```

```python
# convert to pandas df : 
# we got "META deta" that includes page no, movies in it, total pages and results
# the api returns 20 results on each page, so we have in total 241 pages
# we also have 4818 results (movies) 20 of them on each page
pd.DataFrame(data)
```

```python
# what we want is the movies so we create a df or "results":
#and here we can see that we got only page 1
pd.DataFrame(data["results"])
```
