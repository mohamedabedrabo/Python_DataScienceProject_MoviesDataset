### wich column we will use to do the merge ?

```python
# the id column, it's the same column in two tables, but we must analyse it:
credits.id.value_counts()
```

```python
# let's remove duplicates:
credits[credits.duplicated(subset = ["id"], keep = False)].sort_values("id")
```

```python
credits.drop_duplicates(subset = "id", inplace = True)
```

```python
# and now there are no duplicates:
credits.id.value_counts()
```

# Merging the data (Left join)

```python
# left join : the left df is the main df, like SQL
```

```python
# Check the id's that are IN "df" and NOT IN "credits" :
# and there is nothing.
df[~df.id.isin(credits.id)]
```

```python
# Check id's that are IN "credits" and NOT IN "df" :
# there are 741 movies are in credits and are not in df
# these are ppropably the movies that we droped when we cleaned the df
credits[~credits.id.isin(df.id)]
```

```python
# Merge using the function "merge"
df = df.merge(credits, how = "left", left_on = "id", right_on = "id")
```
